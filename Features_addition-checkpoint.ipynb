# Read the datasets

### Previously completed dataset

```python
#labeled dataset for classification
data = pd.read_csv("project_data (1).csv")
data = data.set_index('Countries')
data
```

### New datasets

```python
gender_dev = pd.read_csv("gender_development_index.csv")
gender_eq = pd.read_csv("gender_inequality_index.csv")
enroll_tertiary_edu = pd.read_csv("gross-enrollment-ratio-in-tertiary-education.csv")
literacy = pd.read_csv("literacy-rate-by-country.csv")
school_life = pd.read_csv("school-life-expectancy-from-primary-to-tertiary-education.csv")
cmplt_post_sec = pd.read_csv("share-of-the-population-with-a-completed-post-secondary-education.csv")
cmplt_tertiary = pd.read_csv("share-of-the-population-with-completed-tertiary-education.csv")
human_freedom = pd.read_csv("human-freedom-index-2019.csv")
press_freedom = pd.read_csv("press_freedom_index.csv")
economic_freedom = pd.read_csv("economic_freedom_index.csv")
trust_in_govt = pd.read_csv("share-who-trust-government.csv")
```

# Transform and combine the datasets

```python
#Gross enrollment in tertiary education ratio of population

#selects the maximum value for each country
enroll_tertiary_edu = enroll_tertiary_edu.groupby('Countries').max()
#only keeps the relevant column
enroll_tertiary_edu = enroll_tertiary_edu[['Gross enrolment ratio, tertiary']]
enroll_tertiary_edu
```

```python
#combines with the dataset
mod_data = pd.concat([data, enroll_tertiary_edu],axis=1).dropna()
mod_data
```

```python
#Literacy rate

#selects the maximum value for each country
literacy = literacy.groupby('Country').max()
#only keeps the relevant column
literacy = literacy[['Literacy Rate']]
literacy
```

```python
#combines the datasets
mod_data = pd.concat([mod_data, literacy],axis=1).dropna()
mod_data
```

```python
#School life expectancy

#selects the maximum value for each country
school_life = school_life.groupby("Country")
school_life = school_life.max()
#only keeps the relevant column
school_life = school_life[["School life expectancy, primary to tertiary, both sexes (years)"]]
school_life
```

```python
#Combines the datasets
mod_data = pd.concat([mod_data,school_life],axis=1).dropna()
mod_data
```
